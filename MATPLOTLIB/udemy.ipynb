Matplotlib Axes

```python
plt.plot(numpy_arr2,numpy_arr1, "b*-")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(numpy_arr1,numpy_arr2, "b*-")
plt.subplot(1,2,2)
plt.plot(numpy_arr2, numpy_arr1, "y--")
plt.show()
```

```python
my_figure = plt.figure()
figure_axes = my_figure.add_axes([0.1,0.1,0.7,0.7])
figure_axes.plot(numpy_arr1, numpy_arr2, "r")
figure_axes.set_xlabel("X Axis")
figure_axes.set_ylabel("Y Axis")
figure_axes.set_title("Graph Title")
plt.show()
```

```python
new_fig = plt.figure(dpi=100) #dpi = 1000
new_axes = new_fig.add_axes([0.1,0.1,0.9,0.9])

new_axes.plot(numpy_arr1, numpy_arr1 ** 2 , label = "numpy array ** 2")
new_axes.plot(numpy_arr1, numpy_arr1 ** 3 , label = "numpy array ** 3")
new_axes.legend() 
#new_axes.legend(loc = you can enter number like 1,2,...6, to specify location of numpy array ** 2)
plt.show()
```

```python
new_fig = plt.figure(dpi=1000)
new_axes = new_fig.add_axes([0.1,0.1,0.9,0.9])

new_axes.plot(numpy_arr1, numpy_arr1 ** 2 , label = "numpy array ** 2")
new_axes.plot(numpy_arr1, numpy_arr1 ** 3 , label = "numpy array ** 3")
new_axes.legend(loc = 7) 
plt.show()
```
