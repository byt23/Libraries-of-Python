```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
data = pd.read_csv('athlete_events.csv')
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x = "Height", y="Weight", data = data)
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Athletes Weight vs Height"),
plt.show()
```

```python
sns.set_style("darkgrid")
sns.scatterplot(x = "Height", y="Weight", data = data)
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Athletes Weight vs Height"),
plt.show()
```

```python
sns.set_style("darkgrid")
sns.scatterplot(x = "Height", y="Weight", hue="Sex", data = data)
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Athletes Weight vs Height"),
plt.show()
```

```python
sns.set_style("darkgrid")
sns.scatterplot(x = "Height", y="Weight", hue="Sex", style = "Medal" , size = "Age", data = data)
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Athletes Weight vs Height"),
plt.show()
```

```python
sns.set_style("whitegrid")
sns.set_style("darkgrid")
sns.lineplot(x = "Height", y="Weight", hue="Sex",data = data)
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Athletes Weight vs Height"),
plt.show()
```

```python
sns.set_style("white")
sns.displot(x="Height", data = data)
plt.ylabel("Frequency")
plt.title("Atletes Height Distrbution")
plt.show()
```

```python
sns.set_style("white")
sns.displot(x="Height", data = data, hue= "Sex", kind = "kde")
plt.ylabel("Frequency")
plt.title("Atletes Height Distrbution")
plt.show()
```

```python
sns.barplot(x = "Medal", y="Height", hue = "Sex", data = data)
plt.title("Medals By Height")
plt.show()
```

```python
sns.barplot(x = "Medal", y="Age", hue = "Sex", data = data)
plt.title("Medals By Age")
plt.show()
```

BoxPlot

median, quartile(%25,%50,%75), min/max 1.5 * IQR, outlier

```python
data = np.array([5,7,9,15,20,22,25,30,32,35,37,40,50,55,60,100])

plt.figure(figsize=(6,5))
sns.boxplot(y=data)
plt.title("Box Plot")
plt.ylabel("Data Value")
plt.grid(False)

plt.show()
```

```python
plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
sns.boxplot(x = "class", y = "age" , data = df)
plt.title("Age by Class")
plt.xlabel("Class")
plt.ylabel("Age")


plt.subplot(1,2,2)
sns.boxplot(x = "class", y = "fare" , data = df)
plt.title("Fare by Class")
plt.xlabel("Class")
plt.ylabel("Fare")

plt.tight_layout()
plt.show()
```
