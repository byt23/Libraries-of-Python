Numpy Array Copy vs View

```python
import numpy as np

arr = np.array([1,2,3,4,5])
x = arr.copy()
arr[0] = 42
print(arr)
print(x)
```

```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5, 6, 7])
x = arr.view()
arr[0] = 42
print(arr)
print(x)
```

Every NumPy array has the attribute base that returns None if the array owns the data.

Otherwise, the base attribute refers to the original object.

That's why; The copy returns None.,The view returns the original array.

```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])

x = arr.copy()
y = arr.view()

print(x.base)
print(y.base)
```
