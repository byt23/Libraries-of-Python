z-score = (x - mean) / std

```python
data = np.array([10,12,13,15,18,25,100,105])
```

```python
# Outliers are data points that are significantly different from the rest of the data. 
# They can be identified using z-scores, which measure how many standard deviations a data point is from the mean. 
# A common threshold for identifying outliers is a z-score of 3 or -3,
# Meaning that any data point with a z-score greater than 3 or less than -3 is considered an outlier.
```

```python
mean = np.mean(data)
mean
```

```python
data.sum() / len(data)
```

```python
std = np.std(data)
std
```

```python
z_scores = (data - mean) / std
z_scores
```

```python
data[z_scores > 1]
```
