3- ) Why Do We Use NumPy ?

SIMD Programming = Single Instruction Multiple Data
Numpy is a library for the Python programming language,
adding support for large, multi-dimensional arrays and matrices, 
along with a large collection of high-level mathematical functions to operate on these arrays.

```python
import time

my_list = []

t1 = time.time()
for a in range(1_000_000):
    power = a ** 100
    my_list.append(power)
t2 = time.time()
print(t2 - t1)
```

```python
z3 = time.time()
my_list2 = np.arange(1_000_000) ** 100
z4 = time.time()
print(z4 - z3)
```

Logical Operators and Filtering with NumPy

```python
matrice = np.random.randint(0,10,(5,5))
```

```python
matrice
```

```python
matrice > 5
```

```python
matrice[matrice > 5]
```

```python
matrice[matrice < 5]
```

```python
matrice[matrice == 5]
```

```python
matrice[matrice == 7].size
```

```python
matrice[matrice % 2 == 0]
```

```python
zero = matrice == 0
even = matrice % 2 == 0
```

```python
matrice[even]
```

```python
zero = matrice != 0
one = matrice == 1
even = matrice % 2 == 0
```

```python
matrice[even & zero]
```

```python
matrice[even | one]
```

```python
matrice[(even & zero) | one]
```

```python
matrice.shape
```

```python
m2 = np.full((5,5),5)
```

```python
m2
```

```python
m2[(even & zero) | one]
```
