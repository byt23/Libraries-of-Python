3- ) Why Do We Use NumPy ?

SIMD Programming = Single Instruction Multiple Data
Numpy is a library for the Python programming language,
adding support for large, multi-dimensional arrays and matrices, 
along with a large collection of high-level mathematical functions to operate on these arrays.

```python
import time

my_list = []

t1 = time.time()
for a in range(1_000_000):
    power = a ** 100
    my_list.append(power)
t2 = time.time()
print(t2 - t1)
```

```python
z3 = time.time()
my_list2 = np.arange(1_000_000) ** 100
z4 = time.time()
print(z4 - z3)
```

6-) Changing Data Type In NumPy

```python
numbers = np.array([1, 2, 3, 4, 5, 6])
numbers.dtype
```

```python
numbers = np.array([1, 2.0, 3, 4, 5, 6])
numbers.dtype
```

```python
numbers = numbers.astype("int32")
numbers.dtype
```

```python
numbers = np.array([1, 2, 3, 4, 5, 6], dtype="int8")
numbers.dtype
```

```python
date = np.array("2024-06-01")
date.dtype
```

```python
date = date.astype("datetime64[ns]")
date.dtype
```
